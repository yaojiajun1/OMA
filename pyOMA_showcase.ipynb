```python
X = loadmat("C:/Users/yaoji/Documents/SA/pyoma/data/Run_4_z.mat")  
Data = X["Run_4_y"]
```

use pyOMA.Model to define the initial Condition (example for cov_SSI, you can also try DD_SSI, ITD)
<ul>
<li>Order:</li>
<li>Reference:</li>
<li>Method: you can use different methods including: ITD, cov_SSI, DD_SSI and FDD</li>

Later you may want to plot the Modes. You also need to define the Geometry first.

```python
model = pyOMA.Model(sampling_frequency=12800, Order=50, Data=Data, Method='cov_SSI')
Geometry = np.array([3.1, 12.5, 25, 37.5, 47.5])
```

### Plot Stabilisation Diagramm 

use model.Stabilisation_Diagramm to plot SD: use 
<ul>
<li>svd as indicator with (plot_svd=True)</li>
<li>verbose=True: plot the frequency stable poles and MAC stable poles</li>
<li>verbose=False: plot only the MAC stable poles</li>
</ul>

```python
fig, ax, stable_points, cache = model.Stablisation_Diagramm(plot_svd=True, verbose=True)
```

### choose stable poles
use pyOMA.choose_points to choose points in SD
use left click to choose, and right click to cancel.
when finish, click the middle button.
The chosen points, with their Frequency, damping ration and mode will be save in "Freq_Dr_ESF.dat"
<img width=1200 src="./data/Stabilisation_diagramm.png">  

In order to plot the choosen poles, load "Freq_Dr_ESF.dat" first and use model.plot_Mode() to plot the modes.

```python
freq_dr_esf = np.loadtxt("Freq_Dr_ESF.dat")
model.plot_Mode(freq_dr_esf=freq_dr_esf, Geometry=Geometry)
```
